```python
cpdb2["partner_a"]=""
cpdb2["partner_b"]=""
```

```python
ligids=dict()
for x in liglist:
    ligids[x["from"]]=x["to"]
```

```python
recids=dict()
for x in reclist:
    recids[x["from"]]=x["to"]
```

```python
for x in cpdb2["Ligand"]:
    if x not in ligids:
        ligids[x]="user"
```

```python
cpdb2["partner_a"]=[ligids[x] for x in cpdb2["Ligand"]]
```

```python
for x in cpdb2["Receptor"]:
    if x not in recids:
        recids[x]="user"
```

```python
cpdb2["partner_b"]=[recids[x] for x in cpdb2["Receptor"]]
```

```python
cpdb2
```

```python
cpdb2=cpdb2.loc[cpdb2['partner_a'] != "user"]
cpdb2=cpdb2.loc[cpdb2['partner_b'] != "user"]
```

```python
cpdb2
```

```python
cpdb2.to_csv('inter_ids.csv', index=False)
```
